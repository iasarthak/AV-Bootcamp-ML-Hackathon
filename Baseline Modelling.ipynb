```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import clear_output,display
```

```python
# importing the train and test datasets

train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
sample_sub=pd.read_csv("sample_submission.csv")
```

```python
sample_sub
```

```python
train["Week_No"]=np.ceil(train["Day_No"]/7)
train["Year_No"]=np.ceil(train["Week_No"]/52)
train["Week_No"]=(train["Week_No"]%52)
train.drop(columns="ID",inplace=True)
```

```python
train.loc[train["Week_No"]==0,"Week_No"]=train.loc[train["Week_No"]==0,"Week_No"].apply(lambda x: x+1)
```

```python
test["Week_No"]=np.ceil(test["Day_No"]/7)
test["Year_No"]=np.ceil(test["Week_No"]/52)
test["Week_No"]=(test["Week_No"]%52)
test.drop(columns="ID",inplace=True)
```

```python
test.loc[test["Week_No"]==0,"Week_No"]=test.loc[test["Week_No"]==0,"Week_No"].apply(lambda x: x+1)
```

```python
def export_submission(predictions,filename="Sample.csv"):
    
    ss=sample_sub.copy()
    ss["Sales"]=predictions
    ss.to_csv(f"{filename}.csv",index=False)
```

## Baseline Models

### Mean By Course ID

```python
mean_by_course=train.groupby(["Course_ID"])["Sales"].mean().to_frame()
mean_by_course.reset_index(inplace=True)
mean_by_course
```

```python
test.merge(mean_by_course)
```

```python
export_submission(test.merge(mean_by_course)["Sales"],"Mean_by_Course_ID")
```

### Mean by Course_ID and Week_No

```python
mean_by_course_week=train.groupby(["Course_ID","Week_No"])["Sales"].mean().to_frame()
mean_by_course_week.reset_index(inplace=True)
mean_by_course_week
```

```python
test.merge(mean_by_course_week)
```

```python
export_submission(test.merge(mean_by_course_week)["Sales"],"Mean_by_Course_ID_Week_No")
```

### Mean by Course_ID, Week_No and Short_Promotion

```python
mean_by_course_week_short=train.groupby(["Course_ID","Week_No","Short_Promotion"])["Sales"].mean().to_frame()
mean_by_course_week_short.reset_index(inplace=True)
mean_by_course_week_short
```

```python
test.merge(mean_by_course_week_short)
```

```python
export_submission(test.merge(mean_by_course_week_short)["Sales"],"Mean_by_Course_ID_Week_No_Short_Promo")
```
